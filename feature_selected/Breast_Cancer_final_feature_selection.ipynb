# UPSAMPLING

```python
from sklearn.utils import resample
df_majority=my_data[my_data.diagnosis==0]
df_minority=my_data[my_data.diagnosis==1]
```

```python
df_minority_upsampled=resample(df_minority,replace=True,n_samples=357,random_state=123)
my_data_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
my_data_upsampled.head()
```

```python
my_data_upsampled["diagnosis"].value_counts() 
```

```python
X=my_data_upsampled.drop(['diagnosis'],1)
y=my_data_upsampled['diagnosis']
```

```python
from sklearn.ensemble import ExtraTreesRegressor
selection=ExtraTreesRegressor()
selection.fit(X,y)
```

```python
selection.feature_importances_
```

```python
importance_values=(selection.feature_importances_)*100
col_names_data=X.columns
```

```python
col_names_data
```

```python
dict_data={"feature":col_names_data,"importance":importance_values}
importance_data=pd.DataFrame(dict_data)
importance_data=importance_data.sort_values(by=['importance']) 
importance_data
```

```python
importance_data_features=list(importance_data.feature)
importance_data_features
```

```python
plt.figure(figsize=(15,15))
barlist=plt.barh(list(importance_data["feature"]),list(importance_data["importance"]),color=("hotpink"))

for x in range(0,17):
    barlist[x].set_color('r')

plt.legend(["cutoff<0.5"], loc ="lower right") 
```

```python
X=X.drop(['fractal_dimension_se',
 'radius_se',
 'compactness_mean',
 'smoothness_mean',
 'symmetry_se',
 'smoothness_se',
 'perimeter_se',
 'symmetry_mean',
 'compactness_wors',
 'fractal_dimension_worst',
 'concavity_mean',
 'texture_se',
 'texture_mean',
 'concave points_worst',
 'fractal_dimension_mean',
 'radius_mean',
 'compactness_se'],1)
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

x_transform = scalar.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_transform,y,test_size = 0.30, random_state= 123)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score                #0.9302325581395349
accuracy_score(y_test,y_pred)
```

```python
error_rate=[]
for i in range (1,25):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
error_rate
```

```python
plt.figure(figsize=(10,6))
plt.style.use('ggplot')
plt.plot(range(1,25),error_rate)
plt.xlabel('k')
plt.ylabel("Error Rate")

plt.axvline(x=5,color="blue")
plt.legend(["ERROR","K=21"], loc ="upper right") 
```

```python
y_pred1=model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve
y_pred_proba = knn.predict_proba(X_test)[:,1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
legend_data=roc_auc_score(y_test,y_pred_proba)
legend_data=round(legend_data,4)
legend_data                                            #0.9771          
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
print('FPR')
print(fpr)
print('TPR')
print(tpr)
print('Thresholds')
print(thresholds)
```

```python
plt.style.use('ggplot')
plt.plot(fpr,tpr, label='Knn',color='orange')
plt.plot([0,1],[0,1],'k--',color='navy')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title(' ROC curve')
plt.legend(["AUC=0.9939"], loc ="lower right") 
plt.show()
```
